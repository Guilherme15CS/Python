```python
# Passo 1: Importar a base de dados
import pandas as pd

tabela = pd.read_csv(r'C:\Users\lcamp\OneDrive\Documentos\VS CODE\Python\Hashtag\meus\personality_dataset.csv')

tabela = tabela.dropna()

display(tabela)

tabela.info()
```

```python
# Passo 2: Preparar a base de dados para IA
    # Não pode ter informações como "object", apenas a que deseja ser prevista
    #Stage_fear
    #Drained_after_socializing
    #transformar as colunas de texto em números

from sklearn.preprocessing import LabelEncoder

codificador1 = LabelEncoder()
tabela['Stage_fear'] = codificador1.fit_transform(tabela['Stage_fear'])

codificador2 = LabelEncoder()
tabela['Drained_after_socializing'] = codificador2.fit_transform(tabela['Drained_after_socializing'])

display(tabela)

```

```python
#Separar a base de dados em treino e teste

y = tabela['Personality']
x = tabela.drop(columns='Personality')

#separar em treino e teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
# Passo 3: Criar o modelo de IA -> preve se é introvertido e extrovertido
    #Arvore de decisão
    #Vizinhos mais próximos


#Importar os modelos
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

#Criar os modelos
modelo_arvore = DecisionTreeClassifier()
modelo_vizinhos = KNeighborsClassifier()


#Treinar os modelos
modelo_arvore.fit(x_treino, y_treino)
modelo_vizinhos.fit(x_treino, y_treino)

```

```python
#Passo 4: Escolher o melhor modelo de IA
from sklearn.metrics import accuracy_score


#Fazer previsões
previsao_arvore = modelo_arvore.predict(x_teste)
previsao_vizinhos = modelo_vizinhos.predict(x_teste)


#Verificar a acurácia
display(accuracy_score(y_teste, previsao_arvore))
display(accuracy_score(y_teste, previsao_vizinhos))

#Melhor modelo
    #Vizinhos mais próximos
```
